### Modelling the walls

Each wall will be created from two different layers: materials and isolation. The materials will take 2/3 of the thickness of the wall, while the isolation will require 1/3 of it. The boundaries of the wall are considered to be a simple surface, while the interior will have volume. In the following diagram, the conductances to calculate can be seen:



```python
wall_width = 475 # mm
concrete_width = round(wall_width*2/3/1000,2)
insulation_width = round(wall_width/3/1000,2)

concrete   = {'Conductivity': 1.400,      # W/(mK)
              'Density': 2300.0,          # kg/m
              'Specific heat': 880,       # J/(kgK)
              'Width': concrete_width}    # m

insulation = {'Conductivity': 0.040,      # W/(mK)
              'Density': 16.0,            # kg/m
              'Specific heat': 1210,      # J/(kgK)
              'Width': insulation_width}  # m

wall = pd.DataFrame.from_dict({'Concrete': concrete, 'Insulation': insulation}, orient='index')
wall
```
